```python
from statsmodels.graphics.gofplots import qqplot

def diagnostico(var):
  plt.figure(figsize=(20,4))
  plt.subplot(1,2,1)
  sns.histplot(data = df_red_filtrado,x=var, hue="TP_ESCOLA",kde=True,element='step')
  plt.title(f'Distribuição de {var}')


  plt.subplot(1,2,2)

  sns.boxplot(data=df,x="TP_ESCOLA",y=var)
  plt.title(f'Boxplot de {var}')

  plt.subplot(1,3,2)
  qqplot(df_red_filtrado.loc[df_red_filtrado['TP_ESCOLA']=="Pública",var], line='s',ax=plt.gca())
  plt.title(f'QQ Plot de {var} - Escola Pública')

  plt.tight_layout()
  plt.show()


for var in ['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO','NU_NOTA_MT']:
  diagnostico(var)
```

```python
# teste de hipótese

from scipy.stats import shapiro

for var in ['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO','NU_NOTA_MT']:

  pub = df_red_filtrado.loc[df_red_filtrado['TP_ESCOLA']=="Pública",var].sample(5000,random_state=0)
  priv = df_red_filtrado.loc[df_red_filtrado['TP_ESCOLA']=="Privada",var].sample(5000,random_state=0)

  w1,p1 = shapiro(pub)
  w2,p2 = shapiro(priv)

  print(f'{var}: Pública: {p1:.4f} Privada: {p2:.4f}')
```

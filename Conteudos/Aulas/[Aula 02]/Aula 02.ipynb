```python
import zipfile
import pandas

#zip_file_path = '/content/drive/MyDrive/Mackenzie/MBA ED/DS Experience - Turma F G 1º semestre 2025/Trabalho Final/microdados_enem_2023.zip'
zip_file_path = '/content/drive/MyDrive/[MBA]/[Data Science Experience]/[Documentos]/Trabalho Final/microdados_enem_2023.zip'

csv_path = 'DADOS/MICRODADOS_ENEM_2023.csv'

#descompactar o arquivo
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:  # 'r' - read
    with zip_ref.open(csv_path) as csv_file:
        df = pandas.read_csv(csv_file, sep=';', encoding='latin1')

df.head() #exibe as primeiras 5 linhas
#df.tail() #exibe as últimas 5 linhas

```

```python
df_red = df[['TP_ESCOLA', 'NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO','NU_NOTA_MT']]

df_red.head()
```

```python
#calculando a média na mão (CORNO JOB). Vc não precisa fazer este tipo de cálculo pois já existe função aproriada.
df_red['NU_MEDIA'] = (df_red['NU_NOTA_CN'] + df_red['NU_NOTA_CH'] + df_red['NU_NOTA_LC'] + df_red['NU_NOTA_REDACAO'] + df_red['NU_NOTA_MT']) / 5
df_red.head()
```

```python
df_red.dropna(inplace=True) #exclui os valores nulos - NAN -
df_red.info()
```

```python
df_red['NU_NOTA_MT'].mean()# média
df_red['NU_NOTA_MT'].median()# mediana
```

```python
df_red_filtrado = df_red.loc[df_red['TP_ESCOLA'].isin([2,3])]

df_red_filtrado['TP_ESCOLA'].value_counts()
```

```python
df_red_filtrado.loc[df_red_filtrado['TP_ESCOLA'] == 2,'TP_ESCOLA'] = 'Pública'
df_red_filtrado.loc[df_red_filtrado['TP_ESCOLA'] == 3,'TP_ESCOLA'] = 'Privada'

df_red_filtrado.head()
```

```python
import matplotlib.pyplot as plt

df_red_filtrado.TP_ESCOLA.value_counts().plot(kind='bar')

plt.show()
```

```python
df_red_filtrado[['TP_ESCOLA','NU_NOTA_MT']].groupby('TP_ESCOLA').mean().plot(kind='bar')
plt.show()
```
